Now to remove duplicate rows (with groupby)!

```python
#trying to get a list of just the presidents
for group, frame in df.groupby('Name'):  #can group by party too with df.groupby(['Name', 'Party'])
    print(group, '\n\n') #refers to each one of the unique entrys aka name
    print(frame, '\n\n') #dataframe for each unique name
```

```python
def get_first(group_df): #function called get_first
    return group_df.iloc[0] #get the very first row of each dataframe

df.groupby('Name').apply(get_first) #groupby name and apply get_first   gets first row from each mini dataframe and create a dataframe
#whatever you apply groupby to will be the index. In this case name
```

Now let's reset and answer a few questions about our dataset:
 - Terms in office (activity)
 - Total terms per party

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States")[1]
df = df.rename({'Presidency[a]': 'Number', 'Party[b].1': 'Party', 'President.1': 'Name'}, axis=1)
df = df.drop(['President','Party[b]', 'Presidency[a].1'], axis=1)
df = df.drop(82)
df.head()
```

```python
#way to get number of terms
df.groupby('Name').agg(len) #groupby name and aggergate by length
```

```python
#find number of terms of terms
df['Num_Terms'] = df.groupby('Name')['Number'].transform(len) #group by name than number then transform by length
df.head()
```
