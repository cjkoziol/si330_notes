```python
# Load some sample datasets
import pickle
with open('datasets/dataframes.pkl', 'rb') as f:
    dataframes=pickle.load(f)
```

```python
persons=dataframes['persons']
purchases=dataframes['purchases']
companies=dataframes['companies']
```

```python
display(persons.head())
display(purchases.head())
display(companies.head())
```

```python
#need to figure out the relationship between the dataframes inorder to know what to merge on
#MERGING (by index)
import pandas as pd     #importing pandas
person_df = persons.set_index('credit_card_number')     #creating variable person_df which sets the index of persons to credit_card_number
purchase_df = purchases.set_index('credit_card_number')    #creating variable purchases.df which sets the index of purchases to credit_card_number
```

```python
display(person_df.head())  #display person_df
display(purchase_df.head())  #display purchase_df
pd.merge(person_df, purchase_df, how='inner', left_index = True, right_index = True).head()  #merging person_df and purchase_df and finds where credit card numbers are the same. because you set the index, you need to write left_index = True and right_index = True
```

```python
display(person_df.head())
display(purchases.head())
pd.merge(person_df, purchases, how='inner', left_index = True, right_on = ['credit_card_number']).head()
#because purchases isn't using credit cards as index you specify left_index = True and right on the column credit_card_number
```

```python
import numpy as np

# Question 3
# For everyone who bought something, what's the average price they paid?

df = pd.merge(persons, purchases, how='left', on='credit_card_number') #merging persons and purchases to persons on the credit_card_number
df = df[~df['purchase_id'].isna()] #creating mask and grabbing all the data where isna is not 

results = pd.DataFrame() #new dataframe
for ssn in df['ssn'].unique(): #for each ssn in the unique ssn
    this_person = df[df['ssn'] == ssn] #grab purchases for a particular person
    average = np.mean(this_person['price']) #computes the average for this person
    print(ssn, average) #prints out social security number and corresponding ssn
    
results.head()
```
